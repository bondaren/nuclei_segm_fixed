```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

from scipy import ndimage as ndi

from skimage import measure
from skimage.morphology import watershed
from skimage.feature import peak_local_max
```

```python
def plot_mid(dset, cmap=None, title=None):
    fig = plt.figure(figsize=(10, 10))
    fig.add_subplot(1, 1, 1)
    img = dset[dset.shape[0] // 2, ...]
    if cmap is not None:
        plt.imshow(img, cmap=cmap)
    else:
        plt.imshow(img)
    if title is not None:
        plt.title(title)
```

```python
probs = '/home/adrian/workspace/ilastik-datasets/Vladyslav/vlad_nuclei_inside_predictions.h5'
alpha = 1.
with h5py.File(probs, 'r') as f:
    #background = f['predictions'][0]
    nuclei = f['predictions'][1]
    #result = nuclei - alpha * background
    mask = nuclei > 0.6
    plot_mid(mask)
```
